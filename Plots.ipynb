```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use("seaborn")
```

```python
s0_QNM = pd.read_csv("Data/s0_data.csv")
s0_QNEF = pd.read_csv("Data/s0_QNEF.csv")
s1_QNM = pd.read_csv("Data/s1_data.csv")
s1_QNEF = pd.read_csv("Data/s1_QNEF.csv")
s2_QNM = pd.read_csv("Data/s2_data.csv")
s2_QNEF = pd.read_csv("Data/s2_QNEF.csv")
```

```python
def plott(df, dfQ, s):
    # Create figure for QNM
    #----------------------
    plt.figure()
    plt.title(f"QNM, s = {s}", size = 20)
    plt.xlabel("$\Re(\omega)$", size = 20)
    plt.ylabel("$-\Im(\omega)$", size = 20)
    ax = plt.gca()
    ax.tick_params(axis='both', which='major', labelsize=20)
    
    
    #plot QNMs
    plt.plot(df["Re(w)"][df["l"]==2], -df["Im(w)"][df["l"]==2], "o--", label = "l = 2")
    plt.plot(df["Re(w)"][df["l"]==3], -df["Im(w)"][df["l"]==3], "o--", label = "l = 3")
    plt.plot(df["Re(w)"][df["l"]==4], -df["Im(w)"][df["l"]==4], "o--", label = "l = 4")
    plt.legend(fontsize = 20)
    #----------------------
    # Create figure for QNEF
    #----------------------
    plt.figure()
    plt.title(f"QNEF, s = {s}", size = 20)
    plt.xlabel("$n$ [Overtone number]", size = 20)
    plt.ylabel("$|B_n|$", size = 20)
    plt.xticks([0,1,2,3])
    ax = plt.gca()
    ax.tick_params(axis='both', which='major', labelsize=20)
    #plot QNEFs
    n = [0,1,2,3]
    BN = np.sqrt(dfQ["Re(B)"]**2 + dfQ["Im(B)"]**2)
    plt.plot(n, BN[dfQ["l"]==2], "o--", label = "l = 2")
    plt.plot(n, BN[dfQ["l"]==3], "o--", label = "l = 3")
    plt.plot(n, BN[dfQ["l"]==4], "o--", label = "l = 4")
    plt.legend(fontsize = 20)
```

```python
plott(s0_QNM, s0_QNEF, 0)
plott(s1_QNM, s1_QNEF, 1)
plott(s2_QNM, s2_QNEF, 2)
```
